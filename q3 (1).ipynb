```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

salaries_cleaned = salaries.dropna(subset=['Average salary'])

# Plotting average salary by level of study using a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Level of study', y='Average salary', data=salaries_cleaned)
plt.title('Average Salary by Level of Study')
plt.xlabel('Level of Study')
plt.ylabel('Average Salary (£)')
plt.xticks(rotation=45)
plt.tight_layout()

# Summary statistics: Mean and Median of average salary by level of study
average_salary_stats = salaries_cleaned.groupby('Level of study')['Average salary'].agg(['mean', 'median'])

average_salary_stats
```

```python
response_rates_cleaned = response_rates.dropna(subset=['% Responded'])

# Plotting response rates by graduating year using a line plot for trends over years
grad_years_sorted = response_rates_cleaned['Graduating year'].sort_values().unique()
avg_response_by_year = response_rates_cleaned.groupby('Graduating year')['% Responded'].mean()

plt.figure(figsize=(12, 6))
plt.plot(avg_response_by_year, marker='o', linestyle='-')
plt.title('Average Response Rate by Graduating Year')
plt.xlabel('Graduating Year')
plt.ylabel('Average Response Rate (%)')
plt.xticks(grad_years_sorted, rotation=45)
plt.tight_layout()

# Plotting response rates by level of study using a bar plot
avg_response_by_level = response_rates_cleaned.groupby('Level of study')['% Responded'].mean()

plt.figure(figsize=(10, 6))
avg_response_by_level.plot(kind='bar', color='skyblue')
plt.title('Average Response Rate by Level of Study')
plt.xlabel('Level of Study')
plt.ylabel('Average Response Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()

# Summary statistics: Mean response rate by graduating year and by level of study
response_rate_stats_year = response_rates_cleaned.groupby('Graduating year')['% Responded'].mean()
response_rate_stats_level = response_rates_cleaned.groupby('Level of study')['% Responded'].mean()

response_rate_stats_year, response_rate_stats_level
```
